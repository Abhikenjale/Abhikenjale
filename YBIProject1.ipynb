```python
import cv2
import numpy as np

images = []
for image_path in df.images:
    # Check if image_path is a string before calling cv2.imread
    if isinstance(image_path, str):
        image = cv2.imread(image_path)
        # Check if image loading was successful AND image is not None
        if image is not None and image.size != 0:
            images.append(image)
    else:
        print(f"Warning: Skipping non-string image path: {image_path}")

# Convert the list of images into a NumPy array
images = np.array(images)

# Check if any images were loaded successfully
if images.size == 0:
    print("Error: No images were loaded successfully. Please check image paths and data.")
    # Handle the case where no images were loaded, e.g., exit the script or raise an exception
else:
    # Now you can reshape and create your data variable
    n_samples = len(images)  # Update n_samples based on loaded images
    data = df, images.reshape((n_samples, -1))
```
